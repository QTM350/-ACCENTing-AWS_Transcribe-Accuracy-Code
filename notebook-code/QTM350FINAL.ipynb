```
import pandas as pd
df=pd.read_csv('AllData.csv')
```

```
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```
sns.set(font_scale=2)
viz=sns.lmplot(data=df, x="wrdLength", y="confidence",
    col="accent", hue="accent",x_estimator=np.mean);

plt.xlim([0,14])
plt.xticks(np.arange(0,15,step=2))
plt.subplots_adjust(wspace = 0.3)
```

```
df.groupby(['accent','difficulty']).mean().sort_values(by=['confidence'],ascending=False)
```

```
df.groupby(['accent','difficulty']).std().sort_values(by=['confidence'])
```
